```python
#filter function
def filter_df(df):
    df=df[df.video_url!=df.video_url]
    df=df[df.img_urls == '[]']
    df=df[df.links == '[]']
    return df
```

```python
# drop duplicates
def drop_duplicates(df):
    df=df.drop_duplicates()
    return df
```

```python
#remove redundant columns
def remove_redundant(df):
    df=df.drop(['video_url','links','img_urls','retweets','replies',
                'is_replied', 'is_reply_to', 'reply_to_users'],axis=1)
    return df
```

```python
#features - t_sentiment
def t_sentiment(df):
    df['t_sentiment'] = df['text'].map(lambda x:SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
    return df
    
```

```python
#helper function for converting hashtags to string
def hash_to_string(x):
    w=x[1:-1]
    y=w.split(',')
    z=''
    for st in y:
        z=z+' '+st[1:-1]
    return z
```

```python
#features - h_sentiment
def h_sentiment(df):
    df['h_sentiment'] = df['hashtags'].map(lambda x:SentimentIntensityAnalyzer().polarity_scores(hash_to_string(x))['compound'])
    return df
```

```python
#features - count of punctuations
def p_count(df):
    count = lambda l1,l2: sum([1 for x in l1 if x in l2])
    df['p_count'] = df.text.apply(lambda s: count(s, string.punctuation))
    return df
```

```python
#features - count of hashtags
def h_count(df):
    df['h_count'] = df.hashtags.map(lambda x:len([i.strip() for i in x.split(",")]))
    return df
    
```

```python
#features - readability
def readability(df):
    df['readability'] = df['text'].map(lambda x:ts.automated_readability_index(x))
    return df
    
```

```python
#remove unrequired columns
def remove_unrequired(df):
    df=df.drop(['text','hashtags','likes'],axis=1)
    return df
```

```python
#split the data to test and train
def test_train(df,f=0.2):
    x_train, x_test, y_train, y_test = train_test_split(df.loc[:,'t_sentiment':'readability'], df['like_grade'], test_size=f, random_state=42)
    return x_train,x_test,y_train,y_test
```

```python
#ML models
KNNclassifier = KNeighborsClassifier(n_neighbors = 50, metric = 'minkowski', p = 2)
SVC_model = SVC()
Randomforest = RandomForestClassifier(n_estimators = 300, criterion = 'entropy', random_state = 0)
Decisiontree = tree.DecisionTreeClassifier()
```

```python
df=filter_df(df)
df=drop_duplicates(df)
df=remove_redundant(df)
df=grading(df)
df=t_sentiment(df)
df=h_sentiment(df)
df=p_count(df)
df=h_count(df)
df=readability(df)
df=remove_unrequired(df)
x_train, x_test, y_train, y_test=test_train(df)
```

```python
x_train, x_test, y_train, y_test=test_train(df)
```

```python
#all features
KNNclassifier.fit(x_train.loc[:,'t_sentiment':'readability'],y_train)
SVC_model.fit(x_train.loc[:,'t_sentiment':'readability'],y_train)
Randomforest.fit(x_train.loc[:,'t_sentiment':'readability'],y_train)
Decisiontree.fit(x_train.loc[:,'t_sentiment':'readability'],y_train)

KNN_y_predict = KNNclassifier.predict(x_test.loc[:,'t_sentiment':'readability'])
SVM_y_predict = SVC_model.predict(x_test.loc[:,'t_sentiment':'readability'])
Randomforest_y_predict = Randomforest.predict(x_test.loc[:,'t_sentiment':'readability'])
Decisiontree_y_predict = Decisiontree.predict(x_test.loc[:,'t_sentiment':'readability'])

print('algorithm - accuracy')
print('KNN - ' ,round(accuracy_score(y_test, KNN_y_predict),2))
print('SVM - ' ,round(accuracy_score(y_test, SVM_y_predict),2))
print('Decision Tree - ', round(accuracy_score(y_test, Decisiontree_y_predict),2))
print('Random Forest - ', round(accuracy_score(y_test, Randomforest_y_predict),2))
print('algorithm - classification report')
print('KNN - ', classification_report(y_test, KNN_y_predict))
print('SVM - ', classification_report(y_test, SVM_y_predict))
print('DecisonTree - ', classification_report(y_test, Decisiontree_y_predict))
print('Randomforest - ', classification_report(y_test, Randomforest_y_predict))



```
